```python
import numpy as np
```

```python
n=500
ort=100
std=20
```

```python
alan=np.random.normal(ort,std,(n,1))
```

```python
alan=alan.reshape(-1,1)
alan
```

```python
# y=b0+b1*x
b0=0
b1=2000
fiyat=b0+b1*alan
hata=np.random.normal(0,3000,(n,1))
fiyat+=hata
```

```python
fiyat
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(alan,fiyat)
```

```python
## çözümleme
# verilenler sadece alan ve fiyat istenenler b0 b1 ve hata
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(alan,fiyat)
```

```python
regression.intercept_
```

```python
regression.coef_
```

```python
fiyat_tahmin=regression.predict(alan)
fiyat_tahmin
```

```python
plt.scatter(alan,fiyat)
plt.plot(alan,fiyat_tahmin,color="red")
```

```python
hata_tahmin=fiyat-fiyat_tahmin
hata_tahmin
```

```python
_=plt.hist(hata_tahmin,bins=10)
```

```python
import seaborn as sns
```

```python
sns.distplot(hata_tahmin)
```
