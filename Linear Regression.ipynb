```python
import numpy as np
```

```python
n = 500
average = 100
dev = 20
```

```python
area = np.random.normal(average,dev,(n,1))
```

```python
area = area.reshape(-1,1)
area
```

###  y = b0 + b1 * x

```python
b0 = 0
b1 = 2000
price = b0 + b1 * area
error = np.random.normal(0,3000,(n,1))
price += error
```

```python
price
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(area, price)
```

## Implemention of regressor and prediction

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(area, price)
```

```python
regressor.intercept_ # b0
```

```python
regressor.coef_ # b1
```

```python
pricePrediction = regressor.predict(area)
pricePrediction
```

```python
plt.scatter(area, price)
plt.plot(area, pricePrediction, color="red")
```

```python
errorPrediction = price - pricePrediction
errorPrediction
```

```python
_=plt.hist(errorPrediction,bins=10)
```

```python
import seaborn as sns
```

```python
sns.distplot(errorPrediction)
```

```python
r_area = np.array([80]).reshape(-1,1) # predicted price of 80 square meters house
regressor.predict(r_area)
```
